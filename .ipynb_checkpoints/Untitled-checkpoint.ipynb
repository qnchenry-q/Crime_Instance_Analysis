```python
from sklearn.tree import export_graphviz, DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, precision_score, recall_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.inspection import permutation_importance, plot_partial_dependence

from scipy.special import comb
from scipy import stats


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('feature_df.csv')
df
```

```python
pre_feature_df = df.drop(['unique_key', 'census_tract'], axis = 1)
pre_feature_df = pre_feature_df.rename(columns = {'Unnamed: 0': 'original_idx'}) 
pre_feature_df
```

```python
feature_df = pre_feature_df.copy()
feature_df = feature_df.drop(['original_idx'], axis = 1 )
feature_df
```

```python
#Set up train test split with One HOt encoding
one_hot = OneHotEncoder()
y0 = feature_df.severe

X0 = one_hot.fit_transform(feature_df.copy().drop('severe',axis = 1))
X0_train, X0_test, y0_train, y0_test = train_test_split(X0, y0, 
                                                    test_size=.2)
```

```python
# Train and fit model                                                   
rf0 = RandomForestClassifier(random_state=0,n_jobs=-1)

rf0.fit(X0_train, y0_train)
                                     
# Test Prediction
pred = rf0.predict(X0_test)
print('Accuracy score: {:.3}'.format(rf0.score(X0_test, y0_test)))
```

```python
### Try taking out primary_type since severe is derived from it
y1 = feature_df.severe
X1 = one_hot.fit_transform(feature_df.copy().drop(['severe', 'primary_type'],axis = 1))

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=.2)
```

```python
# Train and fit model                                                   
rf1 = RandomForestClassifier(random_state=0,n_jobs=-1)

rf1.fit(X1_train, y1_train)
                                     
# Test Prediction
pred = rf1.predict(X1_test)
print('Accuracy score: {:.3}'.format(rf1.score(X1_test, y1_test)))
```

```python
feature_df.severe.value_counts()
```

```python
# Confusion Matrix, Gini plot
# Group Data, predict by group

feature_df1 = feature_df.copy().drop(['severe', 'primary_type'], axis = 1)
one_hot_columns = one_hot.get_feature_names(feature_df1.columns)
# Plot the feature importance
feat_scores1 = pd.Series(rf1.feature_importances_, index = one_hot_columns )
feat_scores1 = feat_scores1.sort_values()
ax = feat_scores1[-11:].plot(kind='barh', 
                      figsize=(10,8),
                      color='b')
ax.set_title('Average Gini Importance')
ax.set_xlabel('Average contribution to information gain');
```

```python
cluster_nan_df = target_df_p[target_df_p['cluster'].isna()]
zipcode_nan_df = target_df_p[target_df_p['zipcode'].isna()]
day_name_nan_df = target_df_p[target_df_p['day_name'].isna()]
month_nan_df = target_df_p[target_df_p['month'].isna()]
day_nan_df = target_df_p[target_df_p['day'].isna()]
season_nan_df = target_df_p[target_df_p['season'].isna()]
```

```python
fig, ax = plt.subplots(figsize = (15,15))
props_of_severe_in_nan = {}
for df in important_nans:
    prop = float(len(df[df['severe']==True])/ len(df))
    props_of_severe_in_nan[df.name] = prop
    

total_dist = float(len(target_df_p[target_df_p['severe']==True])/ len(target_df_p))
x_sigma = [x for x in props_of_severe_in_nan.keys()]
x_sigma.append('total data')
y_sigma = [x for x in props_of_severe_in_nan.values()]
y_sigma.append(total_dist)
ax.bar(x_sigma, y_sigma)
ax.set_title('Percentage of Severe crimes in Most Important Features')
print(y_sigma)
```

```python
target_df_s = target_df_p.copy().dropna()
```

```python
#Set up train test split with One HOt encoding
one_hot4 = OneHotEncoder()
y4 = target_df_s.severe

X4 = one_hot4.fit_transform(target_df_s.copy().drop(['severe', 'primary_type'],axis = 1))
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, 
                                                    test_size=.2)
```

```python
# Train and fit model                                                   
rf4 = RandomForestClassifier(random_state=0,n_jobs=-1)

rf4.fit(X4_train, y4_train)
                                     
# Test Prediction
pred = rf4.predict(X4_test)
pred1 = rf4.predict(X4)
print('Accuracy score: {:.3}'.format(rf4.score(X4_test, y4_test)))
print(f'precision: {precision_score(y4, pred1)}')
print(f'recall: {recall_score(y4, pred1)}')
```

```python
feature_df4 = target_df_s.copy().drop(['severe', 'primary_type'], axis = 1)
one_hot_columns4 = one_hot4.get_feature_names(feature_df4.columns)
# Plot the feature importance
feat_scores4 = pd.Series(rf4.feature_importances_, index = one_hot_columns4 )
feat_scores4 = feat_scores4.sort_values()
ax = feat_scores4[-15:].plot(kind='barh', 
                      figsize=(10,8),
                      color='b')
ax.set_title('Average Gini Importance')
ax.set_xlabel('Average contribution to information gain');
```
